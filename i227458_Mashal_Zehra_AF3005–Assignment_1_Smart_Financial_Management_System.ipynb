## **üü¢ Part 1: Loan Eligibility & Interest Rate Calculation [4 marks]**  

üìå **Requirement:**  
SecureBank only approves loans if the **customer meets all eligibility criteria**. The system should:  
‚úî **Check if the customer is employed.**  
‚úî **Verify if the income is at least PKR 50,000.**  
‚úî **Check credit score:**  
   - **750+ ‚Üí 5% Interest Rate**  
   - **650 - 749 ‚Üí 8% Interest Rate**  
   - **Below 650 ‚Üí Loan Rejected**  
‚úî **If the applicant is unemployed, the loan is rejected immediately.**  

### **Implementation Steps:**  
üîπ **Step 1:** Structure the `if-elif-else` statements for this logic.  
üîπ **Step 2:** Implement a Python program using **ipywidgets** for interactive user input (income, credit score, employment status) to decide whether the loan is approved or rejected.

---

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from ipywidgets import interact, widgets


customer_status_widget = widgets.Checkbox(value=False, description='Employed')
customer_income_widget = widgets.IntSlider(min=0, max=100000, step=1000, value=50000, description='Income')
credit_score_widget = widgets.IntSlider(min=300, max=850, step=1, value=650, description='Credit Score')


@interact_manual(customer_status=customer_status_widget, customer_income=customer_income_widget, credit_score=credit_score_widget)
def loan_approval_system(customer_status, customer_income, credit_score):
    if customer_status:
        print("Your loan has been approved!")
        if customer_income >= 50000:
            if credit_score >= 750:
                print("5% interest")
            elif 650 <= credit_score < 750:
                print("8% interest")
            else:
                print("Loan rejected: Credit score under 650")
        else:
            print("Loan rejected: Income under 50000")
    else:
        print("Loan rejected: Customer is unemployed")



```

## **üü¢ Part 2: Investment Risk Assessment [4 marks]**  

üìå **Requirement:**  
SecureBank offers **investment analysis services**. The system should evaluate a **portfolio of stocks** and classify the risk level based on stock returns:  
‚úî If **any stock has a negative return**, the portfolio is **High Risk**.  
‚úî If **all stocks have positive returns**, but at least one return is below 5%, classify as **Medium Risk**.  
‚úî If **all stock returns are 5% or above**, classify as **Low Risk**.  

### **Implementation Steps:**  
üîπ **Step 1:** Use loops to iterate through the list of stock returns.  
üîπ **Step 2:** Implement `if-elif` conditions to classify the risk.  
üîπ **Step 3:** Write a Python program using **ipywidgets** to allow users to **input stock returns interactively** and receive a risk assessment.

---

```python
import ipywidgets as widgets
from IPython.display import display, clear_output


stock_returns = []
output = widgets.Output()


stock_input = widgets.IntText(
    value=0,
    description="Stock Return:",
    style={'description_width': 'initial'}
)


submit_button = widgets.Button(description="Add Return", button_style='success')


done_button = widgets.Button(description="Done", button_style='danger')

def submit_stock(_):
    with output:
        clear_output(wait=True)
        stock_returns.append(stock_input.value)
        print(f"Stock Returns: {stock_returns}")


def finish_input(_):
    with output:
        clear_output(wait=True)
        print(f"Final Stock Returns: {stock_returns}")


        if any(r < 0 for r in stock_returns):
            print("üö® High Risk")
        elif all(r > 0 for r in stock_returns) and any(r < 5 for r in stock_returns):
            print("‚ö†Ô∏è Medium Risk")
        elif all(r > 5 for r in stock_returns):
            print("‚úÖ Low Risk")
        else:
            print("‚ùì Uncategorized Risk")


submit_button.on_click(submit_stock)
done_button.on_click(finish_input)

display(stock_input, submit_button, done_button, output)





```

## **üü¢ Part 3: Loan Repayment Tracker [4 marks]**  

üìå **Requirement:**  
Customers who receive a loan should be able to track their **loan balance** as they make monthly payments. The system should:  
‚úî Start with an **initial loan balance** (e.g., PKR 500,000).  
‚úî Deduct a **fixed monthly payment** (e.g., PKR 25,000).  
‚úî Continue tracking until **loan balance reaches zero**.  
‚úî Display the remaining balance **after each payment**.  

### **Implementation Steps:**  
üîπ **Step 1:** Choose an appropriate loop (`for` or `while`).  
üîπ **Step 2:** Ensure the loop **stops once the loan is fully paid**.  
üîπ **Step 3:** Implement a Python program using **ipywidgets** to **simulate loan repayment interactively**.

---

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
loan_balance = widgets.IntText(
    value=0,
    description="Initial Loan Balance:",
    style={'description_width': 'initial'}
)
submit_button = widgets.Button(description="Add", button_style='success')

output = widgets.Output()
def calculate_loan(_):
  with output:
    clear_output(wait=True)
    balance = loan_balance.value
    f_payment = 25000
    month = 1
    if balance < 25000:
      print("balance should be greater than 25000")
    while balance > 0:
      balance -= f_payment
      print(f"month {month}:the remaining balance is:{balance}")
      month += 1
submit_button.on_click(calculate_loan)
display(loan_balance,submit_button,output)
```


## **üü¢ Part 4: Stock Price Monitoring and Trading Strategy [4 marks]**  

üìå **Requirement:**  
A stock trader wants to **track stock prices daily** and **sell when the price reaches PKR 200**. The system should:  
‚úî Iterate through a **list of stock prices**.  
‚úî **Skip missing stock data** (`None` values).  
‚úî Stop tracking **once the price reaches PKR 200**.  

### **Implementation Steps:**  
üîπ **Step 1:** Handle missing stock data using `continue`.  
üîπ **Step 2:** Stop tracking once the stock hits the target price (`break`).  
üîπ **Step 3:** Write a Python program using **ipywidgets** to **process stock prices interactively** and trigger alerts when conditions are met.

---

```python
import ipywidgets as widgets
from IPython.display import display, clear_output


stock_input = widgets.Textarea(
    value="120, None, 125, 130, None, 140, 160, 180, 200, 220",
    description="Stock Prices (comma-separated):",
    layout=widgets.Layout(width="500px", height="100px"),
    style={'description_width': 'initial'}
)
submit_button = widgets.Button(description="Track Stocks", button_style='success')
output = widgets.Output()


display(stock_input, submit_button, output)


def track_stocks(_):
    with output:
        clear_output(wait=True)


        stock_prices = stock_input.value.split(',')
        stock_prices = [int(price.strip()) if price.strip().lower() != "none" else None for price in stock_prices]

        target_price = 200

        print("Tracking stock prices...\n")

        for price in stock_prices:
            if price is None:
                continue
            print(f"Processing Stock Price: {price}")

            if price >= target_price:
                print(f"üö® Alert! Stock price reached {target_price}. Selling now!")
                break


submit_button.on_click(track_stocks)
```


## **üü¢ Part 5: Currency Exchange Rate Tracker [4 marks]**  

üìå **Requirement:**  
SecureBank provides **real-time currency exchange tracking**. The system should:  
‚úî Start at **PKR 290/USD**.  
‚úî Increase by **1 PKR per day** until it reaches **PKR 300/USD**.  
‚úî Print exchange rates daily and stop when the **target rate is reached**.  

### **Implementation Steps:**  
üîπ **Step 1:** Choose a suitable loop (`for` or `while`).  
üîπ **Step 2:** Stop the loop when the exchange rate reaches the target.  
üîπ **Step 3:** Implement a Python program using **ipywidgets** to **track the currency exchange rate interactively**.

---

```python
import ipywidgets as widgets
from IPython.display import display, clear_output


usd_input = widgets.IntText(value=290, description="Initial USD:", style={'description_width': 'initial'})
pkr_input = widgets.IntText(value=1, description="Daily Increase:", style={'description_width': 'initial'})
submit_button = widgets.Button(description="Start Simulation", button_style='success')


output = widgets.Output()


def simulate_growth(_):
    with output:
        clear_output(wait=True)

        usd = usd_input.value
        pkr = pkr_input.value
        day = 1

        if usd >= 300:
            print("USD is already 300 or more. No simulation needed.")
            return

        while usd < 300:
            usd += pkr
            print(f"Day {day}: USD {usd}")
            day += 1


submit_button.on_click(simulate_growth)


display(usd_input, pkr_input, submit_button, output)



```
